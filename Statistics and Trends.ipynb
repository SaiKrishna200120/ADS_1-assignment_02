Grouping by age

```python
"""Grouping by age using groupby method"""
groupage=social_media.groupby('age')
for age,age_df in groupage:
            print(age)
            print(age_df)
```

```python
groupage.get_group(18)
```

```python
groupsummary= groupage.mean('age')
groupsummary
```

```python
summary=groupsummary.reset_index()
summary
```

Ploting histogram for age, time spent and their income

```python
"""This is histogram frequency plot of age , time spent and their income"""

def histogram():
    social_media.hist(bins=46)
    plt.xlim([0,1000])
    plt.ylim([50,5000])
    #plt.xscale('log')
    plt.yscale('log')
    plt.title("social_media")
    plt.savefig('histogram.png')
    plt.show()
```

```python
histogram()
```

```python
"""This is a bar graph of age versus income in dollar"""
def bar_graph():
    plt.figure(figsize=[10,6])
    plt.bar(summary.age, summary.income_in_dollars, color= 'g')
    plt.xlabel('Age')
    plt.ylabel('income')
    plt.yscale('log')
    plt.title('age vs income')
    plt.savefig('bar_ age vs income.png')
    plt.show()
```

```python
bar_graph()
```

```python
avg_time_genderwise=social_media.groupby(by=['gender']).agg({'time_spent_in_hours':'mean'}).reset_index()
avg_time_genderwise
```

```python
"""This is bar_graph of gender versus average time spent"""
def bar_gender_graph():
    plt.bar(avg_time_genderwise.gender,avg_time_genderwise.time_spent_in_hours,color = 'r')
    plt.yscale('log')
    plt.ylim([0,6])
    plt.xlabel('Gender')
    plt.ylabel('Time spent in hours')
    plt.savefig('bar_gender.png')
    plt.title('gender vs time spent')
    plt.show()
```

```python
bar_gender_graph()
```

Ploting scatter and line graph to visualize the age vs income

```python
"""This is a scatter plot of age versus average time spent in hours"""
def scatter_plot():
    plt.scatter(summary.age, summary.time_spent_in_hours,marker='*')
    plt.plot(summary.age, summary.time_spent_in_hours)
    plt.xlabel("Age Group")
    plt.ylabel('Average Hours Spent')
    plt.yscale('log')
    plt.show()
    plt.savefig('Sactter.png')
```

```python
scatter_plot()
```

Statistical plot

```python
def box_plot():
    dataSet = summary
    print(dataSet)
    figure = plt.figure(figsize =(10, 10))
    plt.boxplot(dataSet,showfliers=False)
    plt.yscale('log')
    plt.show()
    plt.savefig('box.png')
```

```python
box_plot()
```

```python
def violin_plot():
    plt.violinplot(summary, positions=None)
    plt.savefig('violin.png')
    plt.yscale('log')
    plt.show()
```

```python
violin_plot()
```

```python
"""This confusion matrix depicts the likelihood of people owning a car if they were to own a house"""
def confusion_matrix():
    actual = social_media.isHomeOwner
    predicted = social_media.Owns_Car
    confusion_matrix = metrics.confusion_matrix(actual, predicted)
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
    cm_display.plot()
    plt.title('house owner vs car owner')
    plt.show()
    plt.savefig('confusion.png')
```

```python
confusion_matrix()
```
